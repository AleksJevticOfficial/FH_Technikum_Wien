```python
#Prerequisites
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#EXERCISE 2
#1 Load Data and View cols)
df = pd.read_csv('02_dow_jones_index.data')
df
```

```python
#2a - in Pandas, Str = object)
df.select_dtypes("object")
```

```python
#2b) Cleaning Data - Convert to datetime and objects to floats
df['date'] = pd.to_datetime(df['date'])
df['open'] = df['open'].str.strip("$").astype("float64")
df['close'] = df['close'].str.strip("$").astype("float64")
df['high'] = df['high'].str.strip("$").astype("float64")
df['low'] = df['low'].str.strip("$").astype("float64")
df['next_weeks_open']= df['next_weeks_open'].str.strip("$").astype("float64")
df['next_weeks_close'] = df['next_weeks_close'].str.strip("$").astype("float64")
```

```python
df.info()
```

```python
df.describe()
# 7 Magic Numbers: count, mean std, min 25%, 50%, 75%, max
```

```python
#2c) Boxplot and dispersion of numerical values
fig,(ax1,ax2,ax3,ax4)= plt.subplots(ncols=4)
df['open'].plot.box(ax=ax1)
df['close'].plot.box(ax=ax2)
df['high'].plot.box(ax=ax3)
df['low'].plot.box(ax=ax4)
```

```python
#3 plot chart of cisco - CSCO tag)
df_cisco = df.loc[df['stock']=="CSCO",["close","date"]]
df_cisco.sort_values(by=['date'])
plt.plot(df_cisco["date"],df_cisco["close"])
plt.xlabel("Date")
plt.ylabel("price at close/share")
plt.title("Stockprice Cisco")
```

```python
#4) Add a Col with the difference between the highest and lowest price/day
source_data = df[["date","high","low"]]
difference_high_low = []

for i in source_data.iterrows(): #Returns index,Series object
    i = i[1].values
    x = round(i[1]-i[2],2)
    difference_high_low.append(x)
difference_high_low

df['difference_high_low']=difference_high_low
```
